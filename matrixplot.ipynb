```python
import pandas as pd
df = pd.read_csv('2 Cancer_Data.csv')
df
```

```python
df = df.drop('Unnamed: 32', axis=1)
df
```

```python
import pandas as pd
import plotly.express as px
from IPython.display import display
import ipywidgets as widgets


num_cols = df.iloc[:,1:].select_dtypes(include='number').columns.tolist()

checkboxes = {col: widgets.Checkbox(value=False, description=col) for col in num_cols}

for col in num_cols[:3]:
    checkboxes[col].value = True


checkboxes_box = widgets.VBox(list(checkboxes.values()))
display(checkboxes_box)

# Botón para generar el gráfico
button = widgets.Button(description="Generar gráfico")
output = widgets.Output()

def on_button_clicked(b):
    with output:
        output.clear_output()
        
        selected_cols = [col for col, chk in checkboxes.items() if chk.value]
        if len(selected_cols) < 2:
            print("Selecciona al menos 2 columnas.")
            return
        fig = px.scatter_matrix(
            df,
            dimensions=selected_cols,
            color='diagnosis',
            color_discrete_map={'M': '#EF553B', 'B': '#636EFA'},
            title='Características del tumor y diagnóstico',
            height=700
        )
        fig.update_traces(diagonal_visible=False, showupperhalf=False)
        fig.show()

button.on_click(on_button_clicked)
display(button, output)
```
